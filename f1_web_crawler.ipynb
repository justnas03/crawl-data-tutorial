```python
import requests
from bs4 import BeautifulSoup
```

```python
# response is the response from the website
url = 'https://www.formula1.com/en/results/2024/drivers'
response = requests.get(url)
response 
```

```python
# response.content is the HTML content of the webpage
response.content
```

```python
# html.parser is the parser that BeautifulSoup will use to parse the HTML content
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
print(soup.prettify())
```

```python
print(soup.title.text)
```

```python
print(soup.div)
```

```python
print(soup.find('div', class_='modal-content'))
```

```python
# Find the table that contains the data and put it in soup_table
soup_table = soup.find('table', class_='f1-table f1-table-with-data w-full')
# print(soup_table.prettify())

# Find all the rows in the table
soup_table_rows = soup_table.find_all('tr')
# print(soup_table_rows[1].prettify())

for row_id, row in enumerate(soup_table_rows):
    columns = row.find_all('td') #find all the columns in the row
    # print("row = {}: {}".format(row_id, columns))
    if columns: #if the row is not empty
        position = columns[0].text.strip() #strip() is used to remove the whitespace
        driver_name = columns[1].text.strip()[:-3] #.text is used to get the text inside the tag and [: -3] is used to remove the last three characters
        nationality = columns[2].text.strip()
        car = columns[3].text.strip()
        pts = columns[4].text.strip()
        print("position = {},\
              \ndriver_name = {},\
              \nnationality = {},\
              \ncar = {},\
              \npts = {}\n\n".format(position, driver_name, nationality, car, pts))
```
